```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt


Q_default = pd.read_csv('vacuum1-e-1.csv')
Q_breaking = pd.read_csv('vacuum1-e-4.csv')


from scipy.optimize import curve_fit

def func(x, a, b):
    return a * x + b 

popt_d, pcov = curve_fit(func, Q_default['second'], Q_default['Pa'])
print(popt_d)
popt_b, pcov = curve_fit(func, Q_breaking['second'], Q_breaking['Pa'])
print(popt_b)

plt.plot(Q_default['second'],Q_default['Pa'])
plt.plot(Q_default['second'], func(Q_default['second'], *popt_d), 'r-')
# plt.plot(Q_breaking['second'],Q_breaking['Pa'])
# plt.plot(Q_breaking['second'], func(Q_breaking['second'], *popt_b), 'r-')
plt.show()
```
