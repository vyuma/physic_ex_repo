```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import japanize_matplotlib

# データの読み込み
Q_default = pd.read_csv('ベーキング前ガス放出率.csv')
Q_beaking = pd.read_csv('ベーキング後ガス放出率.csv')
from scipy.optimize import curve_fit

def func(x, a, b):
    return a * x + b 

popt_d, pcov = curve_fit(func, Q_default['second'], Q_default['Pa'])
print(popt_d)
popt_b, pcov = curve_fit(func, Q_beaking['second'], Q_beaking['Pa'])
print(popt_b)

plt.plot(Q_default['second'],Q_default['Pa'],label='default')
plt.plot(Q_default['second'], func(Q_default['second'], *popt_d), 'b-',label=f'default fit {popt_d[0].round(3)}')
plt.plot(Q_beaking['second'],Q_beaking['Pa'],label='beaking')
plt.plot(Q_beaking['second'], func(Q_beaking['second'], *popt_b), 'r-',label=f'beaking fit {popt_b[0].round(3)}')
plt.legend()
plt.show()
```

```python
vacuum_up_df = pd.read_csv('ベーキング上昇圧力変化.csv')
vacuum_down_df = pd.read_csv('ベーキング温度降下.csv')


plt.plot(vacuum_up_df['T'],vacuum_up_df['Pa'],label='温度上昇')
plt.plot(vacuum_down_df['T'],vacuum_down_df['Pa'],label='温度下降')
plt.legend(loc='upper left')
plt.show()
```
