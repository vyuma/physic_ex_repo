```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import japanize_matplotlib

A_D = 4.8 * 10**(-2) # [m]
A_T = 7.0 * 10**(-3) # [m]
L = 0.46 # [m]
C_D = 1394 * (2* A_D)**4 / L # [m^3/s]
C_T = 1394 * (2*A_T)**4 / L # [m^3/s]





S = 2 # [L/s]

P = np.linspace(1,1000,100) # [Pa]

def C(C,P):
    return C * P

def S_eff(S,C):
    return S*C / (S+C)

def Thin_S_eff(S,C_1,C_2):
    C = C_1 * C_2 / (C_1 + C_2) 
    return S*C / (S+C)
    

S_D = S_eff(S,C_D*P)
S_T = Thin_S_eff(S,C_T*P,C_D*P)

fig,ax = plt.subplots()
ax.plot(P,S_D*P,label="Default tube")
ax.plot(P,S_T*P,label="Thin tube")
ax.set_xlabel('Pressure [Pa]')
ax.set_ylabel('C[L/s]')
# ax.set_yscale('log')
ax.set_title('C vs P')
ax.legend()
plt.show()

fig,ax = plt.subplots()
ax.plot(P,S_D,label="Default tube")
ax.plot(P,S_T,label="Thin tube")
ax.set_xlabel('Pressure [Pa]')
ax.set_ylabel('Effective flow rate [L/s]')
ax.set_title('実効排気速度')
ax.legend()
plt.show()
```
