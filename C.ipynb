```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt

A_D = 4.8 * 10**(-2) # [m]
A_T = 7.0 * 10**(-3) # [m]
L = 0.46 # [m]

C_D = 1394 * A_D**4 / L # [m^3/s]
C_T = 1394 * A_T**4 / L # [m^3/s]

S = 2 # [L/s]

P = np.linspace(10000,100000,1000) # [Pa]

def S_eff(S,C):
    return S*C / (S+C)

S_D = S_eff(S,C_D*P)
S_T = S_eff(S,C_T*P)

fig,ax = plt.subplots()
ax.plot(P,S_D,label="Dense tube")
ax.plot(P,S_T,label="Thin tube")
ax.set_xlabel('Pressure [Pa]')
ax.set_ylabel('Effective flow rate [L/s]')
ax.legend()
plt.show()
```
